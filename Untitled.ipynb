```python
import json
import os
import csv
import pandas as pd
import numpy as np
```

```python
path = "../Downloads/embeddings/embeddings_f/"
embeddings = []
fnames = []
for file in os.listdir(path):
    with open(path + file) as f:
        for line in f.readlines():
            if line.split('=')[0] in ['llvm-link__main\t', 'llvm-link___QQmain\t', 'FIRModule___QQmain\t']:
                fnames.append(file[:3])
                embeddings.append(line.split('=')[1].split())
            if file[:3] == '602' and line.split('=')[0] == 'llvm-link__cpp_read_main_file\t':
                fnames.append(file[:3])
                embeddings.append(line.split('=')[1].split())


df = pd.DataFrame()
df['fname']=fnames
df['embeddings'] = embeddings
df.to_csv('embeddings/flow-aware-f.csv', index=False)
df.head()
```
